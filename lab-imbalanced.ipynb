Metadata

- **distance_from_home:** the distance from home where the transaction happened.
- **distance_from_last_transaction:** the distance from last transaction happened.

- **online_order:** Is the transaction an online order.
- **fraud:** Is the transaction fraudulent. **0=legit** -  **1=fraud**


```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import precision_score, recall_score, classification_report, confusion_matrix, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.utils import resample
import matplotlib.pyplot as plt
```

```python
fraud = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/card_transdata.csv")
fraud.head()
```

- **1.** What is the distribution of our target variable? Can we say we're dealing with an imbalanced dataset?


```python
fraud['fraud'].value_counts()
```

The dataset is imbalanced, but moderately, not extremely.

- **2.** Train a LogisticRegression.


```python
features = fraud.drop(columns = ["fraud"])
target = fraud["fraud"]

X_train, X_test, y_train, y_test = train_test_split(features, target,random_state=0)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled_np = scaler.transform(X_train)
X_test_scaled_np = scaler.transform(X_test)
```

```python
log_reg = LogisticRegression()
```

```python
X_train_scaled_df = pd.DataFrame(X_train_scaled_np, columns=X_train.columns, index=X_train.index)
X_test_scaled_df  = pd.DataFrame(X_test_scaled_np, columns=X_test.columns, index=X_test.index)
```

```python
log_reg.fit(X_train_scaled_df, y_train)
```

```python
log_reg.score(X_test_scaled_df, y_test)
```

- **3.** Evaluate your model. Take in consideration class importance, and evaluate it by selection the correct metric.


```python
y_pred_test_log = log_reg.predict(X_test_scaled_df)
print(classification_report(y_pred = y_pred_test_log, y_true = y_test))
```

Because the dataset is imbalanced and fraud detection prioritizes identifying fraudulent transactions, accuracy is not an appropriate metric. We therefore focus on recall and F1-score for the fraud class (1). The model achieves a recall of 0.60 and an F1-score of 0.72 for fraud detection.

- **4.** Run **Oversample** in order to balance our target variable and repeat the steps above, now with balanced data. Does it improve the performance of our model? 


```python
train = pd.DataFrame(X_train_scaled_np, columns=X_train.columns, index=X_train.index)
```

```python
train["fraud"] = y_train.values
```

```python
fraudulent = train[train["fraud"] == 1]
non_fraudulent = train[train["fraud"] == 0]
```

```python
len(fraudulent),len(non_fraudulent)
```

```python
yes_oversampled = resample(fraudulent,
                                    replace=True,
                                    n_samples = len(non_fraudulent),
                                    random_state=0)
```

```python
train_over = pd.concat([yes_oversampled, non_fraudulent])
train_over
```

```python
survived_plt = train_over["fraud"].value_counts()
survived_plt.plot(kind="bar")
plt.show()
```

```python
X_train_over = train_over.drop(columns = ["fraud"])
y_train_over = train_over["fraud"]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train_over, y_train_over)
```

```python
y_pred_test_log = log_reg.predict(X_test_scaled_df)
print(classification_report(y_pred = y_pred_test_log, y_true = y_test))
```

After oversampling, the model identifies 95% of fraudulent transactions, compared to only 60% before. However, this comes at the cost of a lower precision, meaning more legitimate transactions are incorrectly flagged as fraud.

- **5.** Now, run **Undersample** in order to balance our target variable and repeat the steps above (1-3), now with balanced data. Does it improve the performance of our model?


```python
train
```

```python
no_undersampled = resample(non_fraudulent,
                                    replace=False,
                                    n_samples = len(fraudulent),
                                    random_state=0)
no_undersampled
```

```python
train_under = pd.concat([no_undersampled, fraudulent])
train_under
```

```python
survived_plt = train_under["fraud"].value_counts()
survived_plt.plot(kind="bar")
plt.show()
```

```python
X_train_under = train_under.drop(columns = ["fraud"])
y_train_under = train_under["fraud"]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train_under, y_train_under)
```

```python
y_pred_test_log = log_reg.predict(X_test_scaled_df)
print(classification_report(y_pred = y_pred_test_log, y_true = y_test))
```

Undersampling transforms the logistic regression model from a conservative fraud detector into a highly sensitive one, substantially reducing false negatives while increasing false positives.

- **6.** Finally, run **SMOTE** in order to balance our target variable and repeat the steps above (1-3), now with balanced data. Does it improve the performance of our model? 

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state = 1,sampling_strategy=1.0)
```

```python
X_train_sm, y_train_sm =  sm.fit_resample(X_train_scaled_df,y_train)
```

```python
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train_sm, y_train_sm)
```

```python
y_pred_test_log = log_reg.predict(X_test_scaled_df)
print(classification_report(y_pred = y_pred_test_log, y_true = y_test))
```

Compared to the baseline logistic regression model, applying SMOTE substantially increases the recall for fraudulent transactions, improving the model’s ability to identify fraud. However, this improvement comes at the cost of reduced precision and overall accuracy due to an increase in false positives. The similar F1-scores indicate that SMOTE primarily shifts the decision boundary rather than improving overall classification performance.
